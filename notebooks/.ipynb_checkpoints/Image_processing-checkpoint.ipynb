```python
import matplotlib.pyplot as plt
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def format_and_render_plot():
    '''Custom function to simplify common formatting operations for exercises. Operations include: 
    1. Turning off axis grids.
    2. Calling `plt.tight_layout` to improve subplot spacing.
    3. Calling `plt.show()` to render plot.'''
    fig = plt.gcf()
    for ax in fig.axes:
        ax.legend(loc='center right')
    plt.show()
```

```python
print('Data type:', jarlid_intact_io.dtype)
print('Min. value:', jarlid_intact_io.min())
print('Max value:', jarlid_intact_io.max())

# Plot the grayscale image
plt.imshow(jarlid_intact_io, vmin=0, vmax=255)
plt.colorbar()
format_and_render_plot()
```

```python
# Import SciPy's "ndimage" module
import scipy.ndimage as ndi

# Generate histogram
hist = ndi.histogram(jarlid_intact_io, min=0, max=255, bins=256)

# Calculate cumulative distribution
cdf = hist.cumsum() / hist.sum()

# Plot the histogram and CDF
fig, axes = plt.subplots(2,1, sharex=True)
axes[0].plot(hist, label='Histogram')
axes[1].plot(cdf, label='CDF')
format_and_render_plot()
```

```python
def show_image_contour(image, contours):
    plt.figure()
    for n, contour in enumerate(contours):
        plt.plot(contour[:, 1], contour[:, 0], linewidth=5, color='green')
    plt.imshow(image, interpolation='nearest', cmap='gray')
    plt.title('Contours')
    plt.axis('off')
    plt.show()
```

```python
# Make the image grayscale
jarlid_grouped = color.rgb2gray(jarlid_grouped)

# Obtain the optimal thresh value
thresh = filters.threshold_otsu(jarlid_grouped)

# Apply thresholding
binary = jarlid_grouped > thresh

# Find contours at a constant value of 0.8
contours = measure.find_contours(binary, 0.8)

# Show the image
show_image_contour(jarlid_grouped, contours)
```

```python
jarlid_grouped = plt.imread('../data/test/p1.jpg')

from skimage.filters import gaussian
smoothed = gaussian(jarlid_grouped, sigma=1)

# Make the image grayscale
smoothed_gray = color.rgb2gray(smoothed)

from skimage.filters import threshold_local

block_size = 35  # tweakable
local_thresh = threshold_local(smoothed_gray, block_size, offset=10)
binary = smoothed_gray > local_thresh

from skimage import morphology

# Remove small objects (noise)
binary = morphology.remove_small_objects(binary, min_size=500)

# Fill holes
binary = morphology.remove_small_holes(binary, area_threshold=500)

# Optional: Dilate to strengthen object edges
binary = morphology.dilation(binary)
```

```python
show_image(binary, "final product")
```

```python
# Find contours at a constant value of 0.8
contours = measure.find_contours(binary, 0.8)

# Show the image
show_image_contour(jarlid_grouped, contours)
```
