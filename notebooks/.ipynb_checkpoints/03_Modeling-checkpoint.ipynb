# 3.3 Loading the datasets

```python
# Loading the datasets Linux system
data_train = np.load("../data/processed/Dataset_365/merged_train.npz")
data_val = np.load("../data/processed/Dataset_365/merged_val.npz")
data_test = np.load("../data/processed/Dataset_365/merged_test.npz")

# Extracting the arrays from the imported data
images_train = data_train['X']
labels_train = data_train['y']

images_val = data_val['X']
labels_val = data_val['y']

images_test = data_test['X']
labels_test = data_test['y']
```

#### A simple CNN structure to test the performance of our dataset

```python
dense_units = 128
batch_size = 256

model_1 = tf.keras.Sequential([
    tf.keras.Input(shape=(128, 128, 1)),
    
    tf.keras.layers.Conv2D(16, 3, padding='same'),
    tf.keras.layers.ReLU(),
    tf.keras.layers.MaxPooling2D(2),
    
    tf.keras.layers.Conv2D(32, 3, padding='same'),
    tf.keras.layers.ReLU(),
    tf.keras.layers.MaxPooling2D(2),
    
    tf.keras.layers.Conv2D(64, 3, padding='same'),
    tf.keras.layers.ReLU(),
    tf.keras.layers.MaxPooling2D(2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(dense_units, activation='relu'),
    
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(2)  # logits
])

optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

model_1.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])

print(f"\nModel Summary")
model_1.summary()

early_stop = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)

history = model_1.fit(
    images_train, labels_train,
    validation_data=(images_val, labels_val),
    epochs=50,
    batch_size=batch_size,
    verbose=1,
    callbacks=[early_stop]
)
```

```python
test_loss, test_acc = model_1.evaluate(images_test, labels_test, verbose=0)
print(f"Accuracy: {test_acc:.4f}")
```

**It seems our augmented dataset is performing quite well. 80% accuracy for a simple CNN structure.**

# 3.4 CNN filter visualization

```python
# Taking a random sample image
idx = random.randint(0, len(images_val) - 1)
sample_image = images_val[idx]
sample_label = labels_val[idx]
```

```python
# Functional API version
inputs = Input(shape=(128, 128, 1))

x = Conv2D(16, 3, padding='same', name='conv2d_1')(inputs)
x = ReLU()(x)
x = MaxPooling2D(2)(x)

x = Conv2D(32, 3, padding='same', name='conv2d_2')(x)
x = ReLU()(x)
x = MaxPooling2D(2)(x)

x = Conv2D(64, 3, padding='same', name='conv2d_3')(x)
x = ReLU()(x)
x = MaxPooling2D(2)(x)

x = Flatten()(x)
x = Dense(dense_units, activation='relu')(x)
x = Dropout(0.3)(x)
outputs = Dense(2)(x)

model_1 = Model(inputs=inputs, outputs=outputs)
```

```python
# Selecting outputs of Conv layers
layer_outputs = [layer.output for layer in model_1.layers if 'conv2d' in layer.name]

'''
What this does:
This is creating a list of outputs from all Conv2D layers in the model, to see what each convolutional layer "sees" — i.e., their activation maps. 
These are the feature maps after filters are applied.

It's like looking inside the brain of the CNN and spying on each layer's reaction to the image.
'''

# Creating a new model that takes the original model's input, but outputs the intermediate Conv layer activations
activation_model = Model(inputs=model_1.input, outputs=layer_outputs)

img_tensor = np.expand_dims(sample_image, axis=0)

'''
The model expects a batch of images, not just a single image.

sample_image has shape (128, 128, 1) (a grayscale image).

But Keras expects shape (batch_size, height, width, channels) — here, (1, 128, 128, 1).

This is essentially faking a batch of size 1. np.expand_dims just adds a new axis at position 0. It’s like wrapping the image in a list:
'''

# Passing an image through
activations = activation_model.predict(img_tensor)

'''
What we get:
A list of NumPy arrays — one for each Conv2D layer.

Each array looks like this: (1, H, W, num_filters)

The first dimension is the batch size (1 here)

These arrays are activation maps — visual fingerprints of what the filters found interesting.
'''
```

```python
def plot_layer_activations(activations, layer_names, images_per_row=8):
    '''A function to plot the activation maps'''
    
    for layer_name, layer_activation in zip(layer_names, activations):
        num_filters = layer_activation.shape[-1]
        size = layer_activation.shape[1]

        n_cols = num_filters // images_per_row  # Number of rows
        display_grid = np.zeros((size * n_cols, images_per_row * size))

        for col in range(n_cols):
            for row in range(images_per_row):
                filter_index = col * images_per_row + row
                if filter_index >= num_filters:
                    break
                channel_image = layer_activation[0, :, :, filter_index]

                # Normalize for visibility
                channel_image -= channel_image.mean()
                if channel_image.std() > 0:
                    channel_image /= channel_image.std()
                channel_image *= 64
                channel_image += 128
                channel_image = np.clip(channel_image, 0, 255).astype('uint8')

                display_grid[
                    col * size: (col + 1) * size,
                    row * size: (row + 1) * size
                ] = channel_image

        # Plot the grid
        scale = 1. / size
        plt.figure(figsize=(scale * display_grid.shape[1],
                            scale * display_grid.shape[0]))
        plt.title(layer_name)
        plt.grid(False)
        plt.imshow(display_grid, aspect='auto', cmap='viridis')
        plt.show()
```

```python
# Getting names of Conv layers only
layer_names = [layer.name for layer in model_1.layers if 'conv2d' in layer.name]

# Plotting activation maps
plot_layer_activations(activations, layer_names)
```

# 3.5 Hyperparemeter tuning

```python
# Defining specific hyperparameter settings 
param_combinations = [
    {'filters': (16, 32, 64),  'kernel_size': 3, 'dense_units': 256, 'dropout_rate': 0.3, 'batch_size': 256, 'learning_rate': 0.001},
    {'filters': (16, 32, 64),  'kernel_size': 3, 'dense_units': 128, 'dropout_rate': 0.3, 'batch_size': 256, 'learning_rate': 0.001},
    {'filters': (16, 32, 64),  'kernel_size': 3, 'dense_units': 256, 'dropout_rate': 0.3, 'batch_size': 128, 'learning_rate': 0.001},
    {'filters': (16, 32, 64),  'kernel_size': 3, 'dense_units': 128, 'dropout_rate': 0.3, 'batch_size': 128, 'learning_rate': 0.001},
    {'filters': (32, 64, 128), 'kernel_size': 3, 'dense_units': 256, 'dropout_rate': 0.3, 'batch_size': 256, 'learning_rate': 0.001},
    {'filters': (32, 64, 128), 'kernel_size': 3, 'dense_units': 256, 'dropout_rate': 0.3, 'batch_size': 128, 'learning_rate': 0.001},

    {'filters': (16, 32, 64),  'kernel_size': 5, 'dense_units': 256, 'dropout_rate': 0.3, 'batch_size': 256, 'learning_rate': 0.001},
    {'filters': (16, 32, 64),  'kernel_size': 5, 'dense_units': 256, 'dropout_rate': 0.3, 'batch_size': 128, 'learning_rate': 0.001},
    {'filters': (32, 64, 128), 'kernel_size': 5, 'dense_units': 256, 'dropout_rate': 0.3, 'batch_size': 256, 'learning_rate': 0.001},
    {'filters': (32, 64, 128), 'kernel_size': 5, 'dense_units': 256, 'dropout_rate': 0.3, 'batch_size': 128, 'learning_rate': 0.001}   
]

results = []

for trial, params in enumerate(param_combinations):
    print(f"\n Trial {trial+1}/{len(param_combinations)}")

    filters = params['filters']
    kernel_size = params['kernel_size']
    dense_units = params['dense_units']
    dropout_rate = params['dropout_rate']
    batch_size = params['batch_size']
    learning_rate = params['learning_rate']

    print(f"Using: Filters={filters}, Kernel={kernel_size}, Dense={dense_units}, Dropout={dropout_rate}, Batch={batch_size}, LR={learning_rate}")

    # Cleaning the memory
    K.clear_session()
    gc.collect()

    model = tf.keras.Sequential([
        tf.keras.Input(shape=(128, 128, 1)),

        tf.keras.layers.Conv2D(filters[0], (kernel_size, kernel_size), padding='same'),
        tf.keras.layers.LeakyReLU(negative_slope=0.1),
        tf.keras.layers.MaxPooling2D((2, 2)),

        tf.keras.layers.Conv2D(filters[1], (kernel_size, kernel_size), padding='same'),
        tf.keras.layers.LeakyReLU(negative_slope=0.1),
        tf.keras.layers.MaxPooling2D((2, 2)),

        tf.keras.layers.Conv2D(filters[2], (kernel_size, kernel_size), padding='same'),
        tf.keras.layers.LeakyReLU(negative_slope=0.1),
        tf.keras.layers.MaxPooling2D((2, 2)),

        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(dense_units, activation='swish'),  # Try relu here in next run
        tf.keras.layers.Dropout(dropout_rate),
        tf.keras.layers.Dense(2)  # Logits output
    ])

    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

    model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])

    print(f"\nModel Summary for Trial {trial+1}")
    model.summary()

    early_stop = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)

    history = model.fit(
        images_train, labels_train,
        validation_data=(images_val, labels_val),
        epochs=30,
        batch_size=batch_size,
        verbose=1,
        callbacks=[early_stop]
    )

    test_loss, test_acc = model.evaluate(images_test, labels_test, verbose=0)
    print(f"Trial {trial+1} - Test Accuracy: {test_acc:.4f}")

    # Save results
    results.append({
        'Trial': trial + 1,
        'Test Accuracy': test_acc,
        'Filters': filters,
        'Kernel Size': kernel_size,
        'Dense Units': dense_units,
        'Dropout Rate': dropout_rate,
        'Batch Size': batch_size,
        'Learning Rate': learning_rate,
        'History': history.history  # Save for plotting curves
    })

    # Extracting only relevant metrics and hyperparameters from results
    df_results = pd.DataFrame([
    {
        'Trial': r['Trial'],
        'Test Accuracy': r['Test Accuracy'],
        'Filters': r['Filters'],
        'Kernel Size': r['Kernel Size'],
        'Dense Units': r['Dense Units'],
        'Dropout Rate': r['Dropout Rate'],
        'Batch Size': r['Batch Size'],
        'Learning Rate': r['Learning Rate']
    }
    for r in results
    ])


    del model
    gc.collect()
```

### Learning Rate Scheduler: CosineDecayRestarts

`CosineDecayRestarts` is a learning rate schedule that gradually **decreases** the learning rate using a **cosine function** and then **restarts** it back to the original value at regular intervals (called **decay cycles**). This is inspired by the SGDR (Stochastic Gradient Descent with Warm Restarts) strategy.

#### Parameters:
- `initial_learning_rate`: The starting learning rate before any decay.
- `first_decay_steps`: Number of steps before the first learning rate reset (defines the length of the first cycle).
- `t_mul`: Factor by which to increase the cycle length after each restart.  
  (e.g., `t_mul=2.0` doubles the cycle duration each time.)
- `m_mul`: Factor to reduce the peak learning rate after each cycle.  
  (e.g., `m_mul=1.0` keeps the peak constant; `m_mul=0.9` reduces it over time.)
- `alpha`: Minimum learning rate fraction (as a float between 0.0 and 1.0).  
  (e.g., `alpha=0.0` decays all the way to zero.)

#### Behavior:

The LR follows a cosine curve down to `alpha * initial_learning_rate`, then jumps back up at the start of each cycle.


#### Use cases

- Helps escape sharp local minima
- Boosts generalization and model robustness
- Excellent for training deep networks (CNNs, Transformers, etc.)
- Adds "wiggle" to the LR schedule to maintain exploration

```python
dense_units = 256
batch_size = 256

model_ed = tf.keras.Sequential([
    tf.keras.Input(shape=(128, 128, 1)),
    
    tf.keras.layers.Conv2D(32, 5, padding='same', kernel_initializer='he_normal'),
    tf.keras.layers.LeakyReLU(negative_slope=0.1),
    tf.keras.layers.MaxPooling2D(2),
    
    tf.keras.layers.Conv2D(64, 5, padding='same', kernel_initializer='he_normal'),
    tf.keras.layers.LeakyReLU(negative_slope=0.1),
    tf.keras.layers.MaxPooling2D(2),
    
    tf.keras.layers.Conv2D(128, 5, padding='same', kernel_initializer='he_normal'),
    tf.keras.layers.LeakyReLU(negative_slope=0.1),
    tf.keras.layers.MaxPooling2D(2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(dense_units, activation='relu', kernel_initializer='he_normal'),
    
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(2)  # logits
])


lr_schedule = tf.keras.optimizers.schedules.CosineDecayRestarts(
    initial_learning_rate=.0001,
    first_decay_steps=1000,
    t_mul=2.0,
    m_mul=1.0,
    alpha=0.0,
    name='SGDRDecay'
)


optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

model_ed.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])

print(f"\nModel Summary")
model_ed.summary()

early_stop = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)

history = model_ed.fit(
    images_train, labels_train,
    validation_data=(images_val, labels_val),
    epochs=50,
    batch_size=batch_size,
    verbose=1,
    callbacks=[early_stop]
)
```

**To choose our best model we will do 3 random runs for each models with pre-defined parameters. All the CNN structure will have the same filter number (32,64,128), dense unit=256 and batch size=256.
```python
# Learning rate scheduler = ReduceLROnPlateau
dense_units = 256
batch_size = 256
num_runs = 3
test_accuracies = []

for trial in range(1, num_runs + 1):
    print(f"\nTrial {trial}")
    
    model_1 = tf.keras.Sequential([
        tf.keras.Input(shape=(128, 128, 1)),
        
        tf.keras.layers.Conv2D(32, 5, padding='same', kernel_initializer='he_normal'),
        tf.keras.layers.LeakyReLU(negative_slope=0.1),
        tf.keras.layers.MaxPooling2D(2),
        
        tf.keras.layers.Conv2D(64, 5, padding='same', kernel_initializer='he_normal'),
        tf.keras.layers.LeakyReLU(negative_slope=0.1),
        tf.keras.layers.MaxPooling2D(2),
        
        tf.keras.layers.Conv2D(128, 5, padding='same', kernel_initializer='he_normal'),
        tf.keras.layers.LeakyReLU(negative_slope=0.1),
        tf.keras.layers.MaxPooling2D(2),
        
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(dense_units, activation='relu', kernel_initializer='he_normal'),
        
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(2)  # logits
    ])

    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
    loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

    model_1.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])

    early_stop = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)

    reduce_lr = ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=3,
        min_lr=1e-6,
        verbose=1
    )

    history = model_1.fit(
        images_train, labels_train,
        validation_data=(images_val, labels_val),
        epochs=50,
        batch_size=batch_size,
        verbose=1,  
        callbacks=[early_stop, reduce_lr]
    )

    # Evaluate on test data
    test_loss, test_acc = model_1.evaluate(images_test, labels_test, verbose=0)
    test_accuracies.append(test_acc)

    print(f"Trial {trial} Test Accuracy: {test_acc:.4f}")

# Show all results
print("\nAll Test Accuracies:")
for i, acc in enumerate(test_accuracies, 1):
    print(f"Run {i}: {acc:.4f}")

print(f"\nAverage Accuracy: {np.mean(test_accuracies):.4f}")
print(f"Standard Deviation: {np.std(test_accuracies):.4f}")
```

## ROC Curve Overview

The **Receiver Operating Characteristic (ROC) curve** plots the **True Positive Rate (TPR)** against the **False Positive Rate (FPR)** at various classification thresholds.  
It’s a graphical representation of a model’s ability to distinguish between the positive and negative classes.

---

### Key Observations:

- **AUC (Area Under Curve) = 0.97**  
  This is **excellent**.  
  - An AUC of **1.0** represents a perfect classifier.  
  - An AUC of **0.5** represents random guessing.  
  An **AUC of 0.97** means the model **almost perfectly separates** the two classes.

- The ROC curve **hugs the top-left corner**, which means:
  - **High True Positive Rate (Recall)**
  - **Low False Positive Rate**
  - **Excellent trade-off** between sensitivity and specificity

- The **dashed diagonal line** represents a **random classifier**.  
  The model's ROC curve is significantly above it, confirming it's **far better than chance**.

---

## Model Implications

### Strengths:
- **Robust at multiple thresholds**  
  The steep initial rise and long plateau suggest the model confidently classifies many positives before significant false positives occur.

- **Generalization power**  
  A high AUC implies good generalization to unseen data.
